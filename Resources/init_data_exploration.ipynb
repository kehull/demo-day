```python
customers = ['zero', 'one', 'three', 'four', 'five', 'six', 'eight', 'nine']
def demographics (x):
    age = round(combined.loc[combined['value'] == x]['age'].dropna().mean())
    income = round(combined.loc[combined['value'] == x]['income'].dropna().mean())
    gender = combined.loc[combined['value'] == x]['gender'].dropna().value_counts(normalize=True)
    desc = combined.loc[combined['value'] == x].describe()
    print(f'customer {x}')
    print(f'average age of customer {x}: {age}')
    print(f'average income of customer {x}: {income}')
    print(gender)
    print(desc)
    print('=====')
for x in customers:
    demographics(x)
```

```python
# if "event" column of row says "offer completed" and "time" column of row == "time" column of row + 1, insert "paid" column
#     value of row +1 into "paid" column of row and delete row +1

i=0

for index, row in combined_3.iterrows():
    if row['event'] == "3 offer completed":
        next_index = index+1
        next_row = combined_3.loc[[next_index]].values.tolist()
        prev_index = index-1
        prev_row = combined_3.loc[[prev_index]].values.tolist()
        next_checker = next_row[-1][1]
        prev_checker = prev_row[-1][1]
        if next_checker == "4 transaction":
            price = next_row[-1][-1]
            combined_3.at[i,'paid'] = price
            i = i+1
        elif prev_checker == "4 transaction":
            price = prev_row[-1][-1]
            combined_3.at[i,'paid'] = price
            i=i+1
        else:
            combined_3.at[i,'paid'] = 0
            i=i+1
    else:
        combined_3.at[i,'paid'] = 0
        i = i+1
combined_3 = combined_3[(combined_3["event"]!="4 transaction")]
combined_4 = combined_3.reset_index(drop=True)
combined_4.head(60)
```
