```python
import pandas as pd
from sklearn.neighbors import DistanceMetric
from scipy.spatial.distance import pdist, squareform
from math import radians
import numpy as np
```

```python
locations = pd.read_csv('coffee_clean.csv') 
locations2 = pd.read_csv('coffee_clean.csv') 
locations
```

```python
x = np.arange(len(locations))
locations['num'] = x
locations['num'] = locations['num'].astype(str)
locations['Stores'] = locations['Store']+locations['num']
locations = locations[['Stores', 'latitude', 'longitude']].rename(columns={'Stores':'Store'})
locations1 = locations[['Store', 'latitude', 'longitude']]
locations1
```

```python
locations['latitude'] = np.radians(locations['latitude'])
locations['longitude'] = np.radians(locations['longitude'])
locations
```

```python
dist = DistanceMetric.get_metric('haversine')
locations[['latitude','longitude']].to_numpy()
```

```python
dist.pairwise(locations[['latitude','longitude']].to_numpy())*3798
```

```python
matrix = pd.DataFrame(dist.pairwise(locations[['latitude','longitude']].to_numpy())*3798,  columns=locations.Store.unique(), index=locations.Store.unique())
matrix
```

```python
matrix = matrix.replace(0, 5)
nearby = matrix[(matrix <= 1).any()]
nearby
```

```python
proximity_list = []
headers = ['store1', 'store2', 'distance']
for index, row in nearby.iterrows():
    for i in range(len(nearby.columns)): 
        if row[i] <=1:
            stores = [row.index[i], index]
            stores_ordered = sorted(stores)
            stores_ordered.append(row[i])
            proximity_list.append(stores_ordered)
proximity = pd.DataFrame(pd.DataFrame(proximity_list, columns=headers)).drop_duplicates()
proximity
```

```python
proximity2 = proximity.merge(locations1, left_on="store1", right_on="Store", how="left")
proximity3 = proximity2.merge(locations1, left_on="store2", right_on="Store", how="left")
proximity4 = proximity3[['store1', 'store2', 'distance', 'latitude_x', 'latitude_y']]
proximity4
locations3 = pd.DataFrame(locations2[['Store', 'latitude']])
proximity5 = proximity4.merge(locations3, left_on="latitude_x", right_on='latitude', how="left").drop('latitude_x', axis=1)
proximity6 = proximity5.merge(locations3, left_on="latitude_y", right_on='latitude', how="left").drop(['latitude_y','latitude_x'], axis=1)
prox_filter = proximity6.loc[proximity6['Store_x'] == "Panera Bread"]
prox_filter2 = prox_filter.loc[prox_filter['Store_y'] == "Starbucks"]
prox_filter3 = prox_filter2.merge(locations1, left_on="store1", right_on="Store", how="left").drop("Store", axis=1)
prox_filter3
```

```python
prox_filter3.to_csv('sbux_panera_conj.csv', index=False)
```
